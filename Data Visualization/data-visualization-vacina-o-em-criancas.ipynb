```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme(style="darkgrid")

print("Setup Complete")
```

```python
vacina_data = '../input/vacina/vacina.csv'

vacina = pd.read_csv(vacina_data, index_col=1, parse_dates=True)

vacina.head(60)
```

```python
vacina = pd.read_csv(vacina_data, index_col=0, parse_dates=True)

plt.figure(figsize=(16,6))

sns.lineplot(data=vacina,  palette=['red'], linewidth=2)
#sns.lineplot(data=vacina,  palette=['yellow'], linewidth=7)
#sns.lineplot(data=vacina,  palette=['lightgreen'], linewidth=4)

plt.title("Ciranças menores de 1 ano vacinadas")
```

```python
vacina = pd.read_csv(vacina_data, index_col=1, parse_dates=True)


plt.figure(figsize=(8,10))

plt.title("Ciranças menores de 1 ano vacinadas")

sns.heatmap(data=vacina, cmap="crest")
```

```python
vacina = pd.read_csv(vacina_data, index_col=0, parse_dates=False)

plt.figure(figsize=(12,8))

plt.title("Ciranças menores de 1 ano vacinadas")

sns.barplot(x=vacina.index, y=vacina['Percentual de cobertura vacinal em menores de 1 ano - Tetravalente'], palette=['#E2498A'])

plt.ylabel("Percentual")
```
